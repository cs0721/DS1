### B.3 Predict for Linear models

```r
# Predict model of y based on X1~4
y = test$Y
y1 = predict(lm(formula = Y ~ X1 + X2 + X3 + X4,data = train),test)
y2 = predict(lm(formula = Y ~ X2 + X3 + X4,data=train),test)

# MSE of Model1
MSE1 = sum((y-y1)^2/length(y))
MSE1
```

```r
# MSE of Model2
MSE2 = sum((y-y2)^2)/length(y)
MSE2
```
